# TCS 

```python
import pandas as pd
df=pd.read_csv('TCS.NS.csv')
print(df.head(5))
print(df.tail(5))
print(df.shape)
```

```python
df[['Date','High','Low']].describe(include='all')
```

# TCS PLOTTING

```python
import matplotlib.pyplot as pp
```

```python
TCS_x = df['Date']
TCS_y = df['High']
TCS_z = df['Low']
pp.figure(figsize=(10,7))
pp.plot(TCS_x, TCS_y, color = "green", label = "Highs")
pp.plot(TCS_x, TCS_z, color = "blue", label = "Lows")
pp.xticks(rotation=90)
pp.title("TCS")
pp.show()
```
